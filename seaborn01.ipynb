Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses

Answer->

five types of plots that can be created using Seaborn library along with their uses:

Line plot: A line plot is used to visualize the trend of a continuous variable over time or any other continuous dimension. It is useful to analyze data that has a clear sequence or time-based structure.

Scatter plot: A scatter plot is used to visualize the relationship between two variables. It is useful to identify patterns or trends in the data and to check for the presence of outliers.

Bar plot: A bar plot is used to compare different categories based on their values. It is useful to visualize the distribution of a categorical variable or to compare different groups of data.

Histogram: A histogram is used to visualize the distribution of a single variable. It is useful to understand the shape of the data and to identify the presence of outliers or unusual values.

Heatmap: A heatmap is used to visualize the relationship between two variables in a tabular format. It is useful to identify patterns and correlations in the data.


```python
import seaborn as sns 
iris = sns.load_dataset('iris')
sns.scatterplot(x = iris.sepal_length , y = iris.sepal_width)
```

```python
sns.lineplot(x=iris.sepal_length,y=iris.sepal_width)
```

```python
sns.barplot(x = iris.sepal_length , y = iris.petal_length)
```

```python
sns.histplot(x="sepal_length", data=iris)
```

```python
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
iris = sns.load_dataset('iris')

corr = iris.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

fmri = sns.load_dataset('fmri')


sns.lineplot(x='timepoint', y='signal', hue='event', style='region', data=fmri)
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.


```python
import seaborn as sns
titanic = sns.load_dataset('titanic')

sns.boxplot(x='pclass', y='age', data=titanic)

sns.boxplot(x='pclass', y='fare', data=titanic)
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
flight = sns.load_dataset('flights')
flight = flight.pivot('month', 'year', 'passengers')
sns.heatmap(data=flight, cmap='coolwarm')

```
